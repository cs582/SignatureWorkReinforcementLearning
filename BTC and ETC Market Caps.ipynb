```python
import json
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
btc = pd.read_csv("historical_data/btc-usd-max.csv", index_col=0).sort_index()['price']
btc.index = [x.split(' ')[0] for x in btc.index.tolist()]

eth = pd.read_csv("historical_data/eth-usd-max.csv", index_col=0).sort_index()['price']
eth.index = [x.split(' ')[0] for x in eth.index.tolist()]
```

```python
plt.figure(figsize=(15,4))
plt.title("Market Cap of BTC and ETH in USD")
plt.plot(btc, label = "BTC", color = 'orange')
plt.plot(eth, label = "ETH", color = 'gray')
plt.ylabel("$USD")
plt.yscale('log')
plt.xticks(btc.index[::500])
plt.legend()
plt.savefig('market_cap.png')
plt.show()
```
